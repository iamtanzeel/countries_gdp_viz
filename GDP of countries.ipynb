```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gdp = pd.read_csv('GDP.csv')
```

```python
gdp.head(50)
```

```python
gdp.dropna(inplace=True)
gdp.drop_duplicates(inplace=True)
```

```python
gdp_top20=gdp.sort_values('Average',ascending=False).head(20)
gdp_top20
```

```python
gdp_india=gdp.loc[gdp['Country']=='India']
gdp_us=gdp.loc[gdp['Country']=='United States']
gdp_uk=gdp.loc[gdp['Country']=='United Kingdom']
gdp_china=gdp.loc[gdp['Country']=='China']
gdp_pakistan=gdp.loc[gdp['Country']=='Pakistan']

nation_gdp=[gdp_india,gdp_china,gdp_pakistan,gdp_uk,gdp_us]
print(nation_gdp)
```

```python
gdp_india=gdp.loc[gdp['Country']=='India']
gdp_us=gdp.loc[gdp['Country']=='United States']
gdp_uk=gdp.loc[gdp['Country']=='United Kingdom']
gdp_china=gdp.loc[gdp['Country']=='China']
gdp_pakistan=gdp.loc[gdp['Country']=='Pakistan']


x_india=gdp_india.columns[1:20]
y_india=gdp_india.iloc[0,1:20]

x_pak = gdp_pakistan.columns[1:20]
y_pak = gdp_pakistan.iloc[0,1:20]

x_china = gdp_china.columns[1:20]
y_china = gdp_china.iloc[0,1:20]

fig=plt.figure()
fig.set_figwidth(30)
fig.set_figheight(30)

plt.plot(x_india,y_india,marker='o',linestyle='dashed',c='orange',mec='black',label='India',lw=5,ms=15)
plt.plot(x_pak,y_pak,marker='o',linestyle='dashed',c='green',mec='black',label='Pakistan',lw=5,ms=15)
plt.plot(x_china,y_china,marker='o',linestyle='dashed',c='red',mec='orange',label='China',lw=5,ms=15)

plt.title('GDP comparison  for India, China and Pakistan (2000-2018)',size=70,c='Blue',pad=110)
plt.xlabel('Year',size=80,labelpad=50)
plt.ylabel('GDP in Billion USD',size=80,labelpad=50)

plt.xticks(fontsize=30,rotation=70)
plt.yticks(fontsize=30)

plt.legend(bbox_to_anchor=(1,1),fontsize=50,facecolor='c',framealpha=0.2,fancybox=True)

plt.ylim((1000,20000))
plt.grid(axis='both',ls='dotted',lw=4)

```

```python
x_india=gdp_india.columns[1:20]
y_india=gdp_india.iloc[0,1:20]

x_pak = gdp_pakistan.columns[1:20]
y_pak = gdp_pakistan.iloc[0,1:20]

x_china = gdp_china.columns[1:20]
y_china = gdp_china.iloc[0,1:20]

fig=plt.figure()
fig.set_figwidth(30)
fig.set_figheight(30)

plt.bar(x_india,y_india,label='India',lw=5,color='orange',alpha=0.9)
plt.bar(x_pak,y_pak,label='Pakistan',lw=5,color='red',alpha=0.3)
plt.bar(x_china,y_china,label='China',lw=5,color='blue',alpha=0.1)

plt.title('GDP comparison for South Asian Major Countires',size=70,c='red',fontweight='bold',pad=100)
plt.xlabel('Year',size=50,labelpad=50)
plt.ylabel('GDP in Billion USD',size=50,labelpad=50)

plt.xticks(fontsize=30,rotation=70)
plt.yticks(fontsize=30)
plt.legend(bbox_to_anchor=(1,1),fontsize=50,facecolor='grey',framealpha=0.2)
plt.ylim((1000,20000))
plt.grid(axis='both',ls='dotted',lw=4)
```

```python
ind_gdp18=gdp_india.iloc[0].at['2018']
us_gdp18=gdp_us.iloc[0].at['2018']
uk_gdp18=gdp_uk.iloc[0].at['2018']
chi_gdp18=gdp_china.iloc[0].at['2018']
pak_gdp18=gdp_pakistan.iloc[0].at['2018']

x=['US','UK','India','China','Pakistan']
y=[us_gdp18,uk_gdp18,ind_gdp18,chi_gdp18,pak_gdp18]
color=['grey','blue','orange','red','green']

fig=plt.figure()
fig.set_figwidth(10)
fig.set_figheight(10)

plt.bar(x,y,color=color)
plt.grid(axis='y',ls='dashed',lw=2)

plt.title('GDP comparison for 2018',color='brown',size=30,fontweight='bold')
plt.xlabel('Countries',size=25,color='purple')
plt.ylabel('GDP in Billion USD',size=25,color='purple')
plt.xticks(x,fontsize=20)
plt.ylim((500,65000))
```

```python
ind_gdp18=gdp_india.iloc[0].at['2018']
us_gdp18=gdp_us.iloc[0].at['2018']
uk_gdp18=gdp_uk.iloc[0].at['2018']
chi_gdp18=gdp_china.iloc[0].at['2018']
pak_gdp18=gdp_pakistan.iloc[0].at['2018']

ind_gdp00=gdp_india.iloc[0].at['2000']
us_gdp00=gdp_us.iloc[0].at['2000']
uk_gdp00=gdp_uk.iloc[0].at['2000']
chi_gdp00=gdp_china.iloc[0].at['2000']
pak_gdp00=gdp_pakistan.iloc[0].at['2000']
                               
ind_percent_growth=round(((ind_gdp18-ind_gdp00)/100),2)
us_percent_growth=round(((us_gdp18-us_gdp00)/100),2)
uk_percent_growth=round(((uk_gdp18-uk_gdp00)/100),2)
chi_percent_growth=round(((chi_gdp18-chi_gdp00)/100),2)
pak_percent_growth=round(((pak_gdp18-pak_gdp00)/100),2)

table={'Countries':['India','US','UK','China','Pakistan'],
              'Percentage growth':[ind_percent_growth,us_percent_growth,uk_percent_growth,chi_percent_growth,pak_percent_growth]}
growth_table=pd.DataFrame(table)
# print(growth_table)
growth_table
```

```python
ind_gdp18=gdp_india.iloc[0].at['2018']
us_gdp18=gdp_us.iloc[0].at['2018']
uk_gdp18=gdp_uk.iloc[0].at['2018']
chi_gdp18=gdp_china.iloc[0].at['2018']
pak_gdp18=gdp_pakistan.iloc[0].at['2018']

ind_gdp00=gdp_india.iloc[0].at['2000']
us_gdp00=gdp_us.iloc[0].at['2000']
uk_gdp00=gdp_uk.iloc[0].at['2000']
chi_gdp00=gdp_china.iloc[0].at['2000']
pak_gdp00=gdp_pakistan.iloc[0].at['2000']
                               
ind_percent_growth=round(((ind_gdp18-ind_gdp00)/100),2)
us_percent_growth=round(((us_gdp18-us_gdp00)/100),2)
uk_percent_growth=round(((uk_gdp18-uk_gdp00)/100),2)
chi_percent_growth=round(((chi_gdp18-chi_gdp00)/100),2)
pak_percent_growth=round(((pak_gdp18-pak_gdp00)/100),2)

x=['US','UK','India','China','Pakistan']
y=[us_percent_growth,uk_percent_growth,ind_percent_growth,chi_percent_growth,pak_percent_growth]
color=['grey','blue','orange','red','green']

fig=plt.figure()
fig.set_figwidth(12)
fig.set_figheight(12)


plt.bar(x,y,color=color)

for i in range(5):
    plt.text(x[i],y[i],y[i],ha='center',va='bottom',size=15)

plt.grid(axis='y',ls='solid')
plt.yticks(np.arange(0,330,30),size=15)
plt.xticks(x,size=15,color='black',alpha=0.8)

plt.title('Percentage GDP growth 2000-2018',fontsize=25,pad=50,color='red')
plt.xlabel('Countries',fontsize=25,color='grey',labelpad=20)
plt.ylabel('Percentage growth',fontsize=25,color='grey',labelpad=20)
plt.show()
```
